```python
# This code defines a reusable function to create a bar chart (countplot)
# that displays the distribution of a categorical variable.
# It then calls that function to display the number of patients for each category in the 'Gender' column.
# A useful feature is that it notes the exact number of patients directly above each bar in the chart.

# patient gender distribution

def univariate_categorical_analysis(var, var_title): 
    plt.figure(figsize=(8, 6)) 
    sns.countplot(x=df[var], palette='viridis') 
    plt.title(f'Distribution of {var_title}', fontsize=16) 
    plt.xlabel(var_title, fontsize=12) 
    plt.ylabel('Quantity', fontsize=12) 

    ax = plt.gca() 
    for p in ax.patches: 
        ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), 
        ha='center', va='center', fontsize=12, color='white', xytext=(0, -15),
        textcoords='offset points')
    plt.show()

# we call the 'gender' column
univariate_categorical_analysis("Gender", "Patient Gender")
```

```python
# This code defines a function called `univariate_analysis` that performs a complete analysis of a single numerical variable.
# For any variable passed to it, the function generates a figure with three side-by-side graphs:
# 1. A boxplot to view outliers.
# 2. A histogram for the frequency distribution.
# 3. A density curve (KDE) for a smoothed view of the distribution.
# Finally, the code calls this function twice to analyze the 'Systolic blood pressure' and 'CK-MB' columns.

def univariate_analysis(var, var_title): 

# Adjust the size and add a general title 
        fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(18, 5)) 
        fig.suptitle(f'Univariate analysis of: {var_title}', fontsize=18, fontweight='bold') 

##boxplot 
        sns.boxplot(data=df, x=var, ax=ax[0], color='cyan') 
        ax[0].set_xlabel(f'Boxplot of {var_title}') 
        ax[0].set_title('Box Plot (Outliers)', fontsize=14) 

## histogram 
        sns.histplot(data=df, x=var, ax=ax[1], kde=True, color='magenta')
        ax[1].set_xlabel(f'Histogram of {var_title}')
        ax[1].set_title('Distribution (Histogram)', fontsize=14)

## density curve
        sns.kdeplot(data=df, x=var, ax=ax[2], fill=True, color='yellow')
        ax[2].set_xlabel(f'Density Curve of {var_title}')
        ax[2].set_title('Density Curve (KDE)', fontsize=14)

        plt.tight_layout(rect=[0, 0, 1, 0.93])
        plt.show()

# first call with the variable 'systolic blood pressure'
univariate_analysis("Systolic blood pressure", "Systolic Blood Pressure")

#second call with the variable 'CK-MB'
univariate_analysis("CK-MB", "CK-MB Level")
```

```python
# This code calculates the correlation matrix between all numeric columns in the DataFrame and displays it as a heatmap.
# First, it automatically selects only columns that contain numbers.
# Then, it calculates the strength of the linear relationship between each pair of these columns.
# Finally, it draws a heatmap where the color of each cell represents the strength of the correlation, and the exact number (formatted to two decimal places) is displayed inside the cell.

plt.figure(figsize=(8, 6))
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = df[numeric_cols].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='viridis', fmt=".2f")
plt.title('Heart Attack correlation')
```

```python
# This code defines a very comprehensive function called `bivariate_analysis` to compare a numerical variable
# against a categorical variable (the diagnosis).
# The function generates three graphs at once (histogram, density, and boxplot) to provide a detailed view.
# The code then attempts to call this function four times to analyze different variables such as age,
# blood sugar, heart rate, and CK-MB, comparing them with the diagnosis result.

def bivariate_analysis(dataframe, var_x, var_hue, var_x_title):
    fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(20, 6))

#Histogram
    sns.histplot(data=dataframe, x=var_x, hue=var_hue, ax=ax[0], palette='viridis')
    ax[0].set_title(f'Histogram of {var_x_title}')
    ax[0].set_xlabel(var_x_title)
    ax[0].set_ylabel('Frequency')
    ax[0].legend(title='Diagnosis', labels=['More likely', 'Less likely'])

#Density Chart (KDE) 
    sns.kdeplot(data=dataframe, x=var_x, hue=var_hue, ax=ax[1], fill=True, palette='viridis') 
    ax[1].set_title(f'Density of {var_x_title}') 
    ax[1].set_xlabel(var_x_title) 
    ax[1].set_ylabel('Density') 
    ax[1].legend(title='Diagnosis', labels=['More likely', 'Less likely']) 


#Boxplot 
    sns.boxplot(data=dataframe, x=var_hue, y=var_x, ax=ax[2], palette='viridis') 
    ax[2].set_title(f'Boxplot of {var_x_title}') 
    ax[2].set_xlabel('Diagnosis (0 = Less, 1 = More likely)')
    ax[2].set_ylabel(var_x_title)

    plt.suptitle(f'Bivariate Analysis of "{var_x_title}" vs. Heart Attack Diagnosis', fontsize=16)
    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.show()

bivariate_analysis(df, 'Age', 'Result', 'Patient Age')

bivariate_analysis(df, 'Blood sugar', 'Result', 'Blood sugar level')

bivariate_analysis(df, 'Heart rate', 'Result', 'By heart rate')

bivariate_analysis(df, 'CK-MB', 'Result', 'by CK-MB')
```

##### Checking and handling outliers

```python
# This code uses the Interquartile Range (IQR) method, a standard statistical technique, to detect outliers in the 'Age' column.
# 1. Calculate the first quartile (Q1, the value below which 25% of the data fall) and the third quartile (Q3, 75%).
# 2. Calculate the Interquartile Range (IQR = Q3 - Q1).
# 3. Define the lower and upper bounds. Any value that is 1.5 times the IQR below Q1 or above Q3 is considered an outlier.
# 4. Finally, filter and display the rows in the DataFrame that contain these outliers in the 'Age' column.

# For the 'Age' column
Q1_age = df['Age'].quantile(0.25)
Q3_age = df['Age'].quantile(0.75)
IQR_age = Q3_age - Q1_age
lower_bound_age = Q1_age - 1.5 * IQR_age
upper_bound_age = Q3_age + 1.5 * IQR_age
outliers_age = df[(df['Age'] < lower_bound_age) | (df['Age'] > upper_bound_age)]
print(f"Number of outliers in 'Age': {len(outliers_age)}")
outliers_age
```

```python
# Checking outliers for CK-MB
Q1_ck = df['CK-MB'].quantile(0.25)
Q3_ck = df['CK-MB'].quantile(0.75)
IQR_ck = Q3_ck - Q1_ck
outliers_ck = df[(df['CK-MB'] < (Q1_ck - 1.5 * IQR_ck)) | (df['CK-MB'] > (Q3_ck + 1.5 * IQR_ck))]
print(f"Number of outliers in 'CK-MB': {len(outliers_ck)}")
outliers_ck
```

```python
#replace values ​​greater than 50
df.loc[df['CK-MB'] > 50, 'CK-MB'] = 50
```

### **Decision Tree model implementation with hyperparameter tuning**

#### Import Libraries

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
# This code block prepares the data to train the model
# 1. X = df.drop("Result", axis=1): Creates a new DataFrame 'X' containing all the columns from the original DataFrame 'df' EXCEPT the "Result" column. 'X' represents the features that will be used for prediction.
# 2. y = df["Result"]: Creates a Series 'y' containing ONLY the "Result" column. 'y' is the target variable, that is, what the model will attempt to predict.
# 3. X.head(): Displays the first 5 rows of the DataFrame 'X' to verify that the "Result" column was successfully dropped.

X = df.drop("Result", axis=1)
y = df["Result"]
X.head()
```

```python
# This line of code splits the dataset into two subsets: one for training and one for testing.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
# This code block uses GridSearchCV to automatically find the best hyperparameter combination for a Decision Tree model.
# 1. `param_grid`: Defines a dictionary with all the hyperparameter values ​​to be tested (tree depth, minimum samples, etc.).
# 2. `GridSearchCV`: Configures the search tool. It will systematically test each `param_grid` combination, using cross-validation (cv=5) to robustly evaluate each one.
# 3. `grid_search.fit()`: Starts the search and training process on the data. This is the part that does all the heavy lifting.
# 4. `best_tree`: Once the search is complete, this line extracts and saves the model (the tree) with the configuration that obtained the best score ('accuracy').
from sklearn.model_selection import GridSearchCV
param_grid = {
    "max_depth": [2, 4, 6, 8, 10],
    "min_samples_split": [2, 3, 4, 5, 10],
    "min_samples_leaf": [1, 2, 3, 4, 5],
    "criterion" : ["gini", "entropy"]
}

grid_search = GridSearchCV(DecisionTreeClassifier(random_state=42),
                           param_grid,
                           cv=5,
                           scoring="accuracy"
                           )

grid_search.fit(X_train, y_train)

best_tree = grid_search.best_estimator_
```

```python
# This line of code uses the best decision tree (`best_tree`), which was optimized in the previous step, to make predictions on the training dataset (`X_train`).
y_pred_hyper = best_tree.predict(X_train)
```

```python
best_tree
```

```python
# This line of code calculates the accuracy score of the model on the training data.
accuracy_score(y_train, y_pred_hyper)
```

```python
# This line of code creates a visualization called a "heatmap" for the model's confusion matrix on the training data.
# 1. `confusion_matrix(y_train, y_pred_hyper)`: First, it calculates the confusion matrix by comparing the actual labels (`y_train`) with the model's predictions (`y_pred_hyper`). This 2x2 matrix shows a detailed summary of the hits and misses.
# 2. `sns.heatmap(...)`: Then, it draws that matrix as a heatmap. The color of each cell represents the number of predictions.
sns.heatmap(confusion_matrix(y_train, y_pred_hyper), annot=True, cmap="Blues", fmt="d")
```

```python
# This block of code evaluates the performance of the optimized model on completely new data.
y_pred_hyper_test = best_tree.predict(X_test)
accuracy_score(y_test, y_pred_hyper_test)
```

```python
# This code generates a heatmap to visualize the model's confusion matrix, but this time on the test data set.
# 1. `confusion_matrix(y_test, y_pred_hyper_test)`: Calculates the confusion matrix by comparing the actual results from the test set (`y_test`) with the predictions the model made for that data (`y_pred_hyper_test`).
# 2. `sns.heatmap(...)`: Draws the matrix as a heatmap. This visualization is crucial for understanding the model's actual performance on data it hasn't seen before.
sns.heatmap(confusion_matrix(y_test, y_pred_hyper_test), annot=True, cmap="Blues", fmt="d")
```

```python
# This code block standardizes the dataset's features, a common and often necessary preprocessing step for models like Logistic Regression.
# 1. `scaler = StandardScaler()`: Creates a "scaler" object that will learn to normalize the data.
# 2. `X_Scaled = scaler.fit_transform(...)`: Calculates the mean and standard deviation of each feature column and then transforms them so that they all have a mean of 0 and a standard deviation of 1. This prevents variables with larger ranges from dominating the model.
# 3. `X_Scaled_df = pd.DataFrame(...)`: Converts the scaled data (which is in an array format) back to a pandas DataFrame, preserving the original column names.
# 4. `X_Scaled_df`: Displays the new DataFrame with all its features at the same scale.
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_Scaled = scaler.fit_transform(df.drop("Result", axis=1))

X_Scaled_df = pd.DataFrame(X_Scaled, columns=df.drop("Result", axis=1).columns)
X_Scaled_df
```

```python
X_Scaled_df.hist(figsize=(10,10))
```

```python
# This line of code splits the already scaled data (X) and the target variable (y) into training and test sets.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# This block of code prepares and trains the Logistic Regression model.
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
model = LogisticRegression()

model.fit(X_train, y_train)
```

```python
# This line of code uses the already trained Logistic Regression model (`model`) to make predictions on the same dataset it was trained on (`X_train`).
y_pred = model.predict(X_train)
```

```python
# This line calculates the accuracy score of the model on the training data.
accuracy_score(y_train, y_pred)
```

```python
# This code generates a heatmap to visualize the model's confusion matrix on the training data.
# 1. `confusion_matrix(y_train, y_pred)`: Calculates the matrix that summarizes the model's performance, comparing the predictions (`y_pred`) with the actual values ​​(`y_train`).
# 2. `sns.heatmap(...)`: Draws this matrix as a visual graph.
sns.heatmap(confusion_matrix(y_train, y_pred), annot=True,cmap="Blues", fmt="d")
```

```python
# This block of code measures the final performance of the model on data it has never seen before.
y_pred_test = model.predict(X_test)
accuracy_score(y_test, y_pred_test)
```

```python
# This code generates a heat map for the confusion matrix, evaluating the model's performance on the test set.
# 1. `confusion_matrix(y_test, y_pred_test)`: Calculates the matrix by comparing the actual test results (`y_test`) with the model's predictions for that data (`y_pred_test`).
# 2. `sns.heatmap(...)`: Draws the matrix as a heat map for easy visual interpretation of hits and misses (True Positives/Negatives and False Positives/Negatives).
sns.heatmap(confusion_matrix(y_test, y_pred_test), annot=True,cmap="Blues", fmt="d")
```
